# **Best Goalkeepers**

```python
jk=['saved','saved_penalties','cleansheets','punches made']
```

```python
df_goalkeeper=df_goalkeeping[['player_name','club']+jk]
df_goalkeeper.head()
```

```python
df_goalkeeper.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler=MinMaxScaler()
df_goalkeeper[jk]=Scaler.fit_transform(df_goalkeeper[jk])
```

```python
df_goalkeeper.head()
```

```python
df_goalkeeper['goalkeeper_score']=0.5*df_goalkeeper['saved']+0.1*df_goalkeeper['saved_penalties']+0.3*df_goalkeeper['cleansheets']+0.1*df_goalkeeper['punches made']
```

```python
g=df_goalkeeper.sort_values(by='goalkeeper_score',ascending=False).head(10)
g
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='player_name',y='goalkeeper_score',data=g)
plt.xticks(rotation=45)
plt.title('Top 10 Goalkeepers of the UCL 2021/22',fontsize=16,weight='bold')
plt.xlabel('')
plt.ylabel('Goalkeeper score')
for i,v in enumerate(g['goalkeeper_score']):
  plt.text(i,v+0.01,round(v,2),ha='center')
plt.tight_layout()
plt.savefig('best_goalkeeper.jpg',bbox_inches='tight')
plt.show()
```

# Removing outliers

```python
sns.boxplot(y='goals',data=df_goals)
```

```python
q1=df_goals['goals'].quantile(0.25)
q3=df_goals['goals'].quantile(0.75)
iqr=q3-q1
```

```python
lower=q1-(1.5*iqr)
upper=q3+(1.5*iqr)
print(lower,upper)
```

```python
df_clean=df_goals[(df_goals['goals']>=lower)&(df_goals['goals']<=upper)]
df_clean
```

```python
sns.boxplot(y='goals',data=df_clean)
```
